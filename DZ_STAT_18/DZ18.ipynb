Монте карло

```python
trials = 1000000
successes = 0
summ = 0

for i in range(trials):
    rnd = str(np.random.randint(0, 1000))
    if rnd in '777':
        summ+=200
        successes+=1
    elif rnd in '999':
        summ+=100
        successes+=1
    elif rnd in '555':
        summ+=50
        successes+=1
    elif rnd in '333':
        summ+=15
        successes+=1
    elif rnd in '111':
        summ+=10
        successes+=1
    elif rnd[1:] in '77':
        summ+=5
        successes+=1
    elif rnd[1:] in '00':
        summ+=2
        successes+=1
    elif rnd[-1] in '7':
        summ+=3
        successes+=1
    elif rnd[-1] in '0':
        summ+=1  
        successes+=1

print('P=',successes / trials)  
print(round((summ/trials), 2))
```
