```python
def region_detect(phrase):
    geo_data = {

    'Центр': ['москва', 'тула', 'ярославль'],

    'Северо-Запад': ['петербург', 'псков', 'мурманск'],

    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
    
    words = phrase.split(' ')
    
    for word in words:
        if word in sum(geo_data.values(),[]):
            for reg,cit in geo_data.items():
                if word in cit:
                    return reg
        return 'undefined'
    
```

```python
geo['region'] = geo['keyword'].apply(region_detect)
```

```python
def years_func(m,n):
    """
    Создаем диапазон дат
    """
    years = []
    i = m
    while i <= n:
        years.append(str(i))
        i += 1
    return years
```

```python
years = years_func(1950,2010)
```

```python
import re
```

Напишите функцию production_year, которая каждой строке из названия фильма выставляет год выпуска. Не все названия фильмов содержат год выпуска в одинаковом формате, поэтому используйте следующий алгоритм:

для каждой строки пройдите по всем годам списка years
если номер года присутствует в названии фильма, то функция возвращает этот год как год выпуска
если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год


```python
def production_year(year_row):
    #вытаскиваем год всех фильмов, у которых есть год в названии. Если года нет возвращаем 1900
    try:
        no_space = re.sub(r'\s$','',year_row)
        ye = re.search(r'(19|20|18)\d\d(\)|)$', no_space)
        year = re.sub(r'\)','',ye.group(0))
        return year
    except AttributeError:
        return '1900'
```

```python
title['year'] = title['title'].apply(production_year)
```

```python
title.sort_values(by='year')
```

Посчитайте средний рейтинг всех фильмов для каждого значения столбца ‘year’ и отсортируйте результат по убыванию рейтинга

```python
title_years = title[title['year'].isin(years) == True] # Выводим фильмы с 1950 по 2010
```

```python
# Джойним таблицу title_years с таблицей с рейтингами ratings
title_and_rate = title_years.merge(ratings, on='movieId', how ='inner') 
```

```python
pd.options.display.max_rows = 100 # увеличиваем кол-во строк на вывод до 100
```

```python
#Выводим рейтинг всех фильмов с группировкой по годам и по убыванию рейтинга
title_and_rate[['year','rating']].groupby(by='year').mean().sort_values(by='rating', ascending = False)
```
