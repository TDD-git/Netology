## Задание 2

```python
data = pd.read_csv('power.csv')
data.head()
```

```python
prib = data[ ((data['country']=='Lithuania') | (data['country']=='Latvia') | (data['country']=='Estonia'))]
prib.head()
```

```python
summ_cons = prib[ ((2005 < prib['year']) & (prib['year'] < 2010)) & ((prib['category'] == 4) | (prib['category'] == 12) | (prib['category'] == 21))]
total = summ_cons['quantity'].sum()
print(f'Суммарное потребление стран Прибалтики: {total}')
```
