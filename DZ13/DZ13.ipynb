## Задание 1

```python
log = pd.read_csv('visit_log.csv',sep=';')
```

```python
#если источник traffic_source равен yandex или google, то в source_type ставится organic
log.loc[log.traffic_source.isin(['yandex', 'google']), 'source_type'] = 'organic'

#для источников paid и email из России - ставим ad
log.loc[(log.traffic_source.
        isin(['paid', 'email'])) 
        & (log.region == 'Russia'),'source_type'] = 'ad'

#для источников paid и email не из России - ставим other
log.loc[(log.traffic_source.
        isin(['paid', 'email'])) 
        & (log.region != 'Russia'),'source_type'] = 'other'

#все остальные варианты берем из traffic_source без изменений
log['source_type'] = log['source_type'].fillna(log['traffic_source'])
```

```python
from pymystem3 import Mystem
```

```python
data = pd.DataFrame({
    'keyword': ['курс гривны к рублю', 'доллары в рубли', '100 долларов в рублях', 'курс рубля'],
    'shows': [125076, 114173, 97534, 53546],
})
```

```python
def lemmas(row):
    m = Mystem()
    lemmas = m.lemmatize(row)
    return ' '.join(lemmas[0:-1]) #переводим в строку без последнего символа элемента перевода строки \n
```

```python
data['lemmas'] = data['keyword'].apply(lemmas)
```
