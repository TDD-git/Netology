```python

Headers = []
with open('Header.txt', 'r') as p:
    for row in p:
        ye = re.search(r'\s+\b(\D+(\d|))', row)
        header = ye.group(1)
        Headers.append(header.strip())
     
```

## Задание 2

Посчитаем выбросы по основным аттрибутам

```python
horses.describe()
```

```python
# Найдем выбросы через межквартильный размах
def outliers(column):
    q1 = horses[column].quantile(0.25)
    q3 = horses[column].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - (1.5 * iqr) 
    upper_bound = q3 + (1.5 * iqr)
    remove_outliers = horses[horses[column].between(lower_bound, upper_bound, inclusive=True)]
    return remove_outliers
```

```python
outliers('rectal temperature')
```

```python
outliers('rectal temperature')
```

```python
horses.describe()
```

```python
outliers('respiratory rate')
```

```python
outliers('pulse')
```

Как видно что довольно много выбросов у показателей частота дыхания и температура. По имеющимся данным мы не можем четко сказать что выбросы были ошибочный, т.к. например пульс лошади может зависеть от породы и ее предназначения(во время забегов у некоторых может быть больше 250), так же и частота дыхания и температура. Поэтому нет причин их не учитывать.

Как видно выше, в столбцах 'rectal temperature', 'respiratory rate' и 'pain' достаточно много пропусков. Но так как и мода и медиана и среднее по температуре находятся довольно близко, то я решил заменить недостающие значения с помощью среднего

```python
horses_na = pd.read_csv('horse_data.csv',names=Headers, na_values='?')
horses_na_eight = horses_na.iloc[:,[0,1,3,4,5,10,-6,2]]
horses_na_no_dubl = horses_na_eight.drop_duplicates(subset = 'Hospital Number',  keep='first')
```

```python
"""
Так как значение показателей моды, медианы и среднего были довольно близки, то заполнил пропуски значением среднего.

"""
horses_na_no_dubl['rectal temperature'].describe()
```

```python
horses_na_no_dubl['rectal temperature'].mode()[0]
```

```python
horses_na_no_dubl.iloc[:, 2].fillna(horses['rectal temperature'].mean(), inplace = True)
```

```python
"""
С пульсом оказалось все сложнее, был очень большой разброс,
поэтому пришлось группировать по Age и outcome и заполнять медианными значениями.
"""
horses_na_no_dubl.pulse.describe()
```

```python
horses_na_no_dubl.pulse.mode()[0]
```

```python
horses_na_no_dubl.iloc[:, 3].fillna(horses.groupby(['Age','outcome']).pulse.transform('median'), inplace=True)
```

```python
"""
C уровнем боли оказлось все проще, и мода медиана и среднее оказлись ~ 3
"""
horses_na_no_dubl.pain.describe()
```

```python
horses_na_no_dubl.pain.mode()[0]
```

```python
horses_na_no_dubl.iloc[:, 5].fillna(horses.pain.mode()[0], inplace = True)
```

```python
"""
Частоту дыхания заполняем так же в разбивки по возрасту и исходу
"""
```

```python
horses_na_no_dubl['respiratory rate'].describe()
```

```python
horses_na_no_dubl['respiratory rate'].mode()[0]
```

```python
horses_na_no_dubl.iloc[:, 4].fillna(horses.groupby(['Age','outcome'])['respiratory rate'].transform('median'), inplace=True)
```
